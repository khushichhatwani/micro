```python
import numpy as np
import pandas as pd
import sklearn.datasets as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_iris()
```

```python
print(iris.DESCR)
```

```python
iris.feature_names
```

```python
iris.target
```

```python
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
df['Iris type'] = iris['target']
```

```python
df
```

```python
def convert(x):
    if x == 0:
        val = 'setosa'
    elif x == 1:
        val = 'versicolor'
    else:
        val = 'virginica'
    return val
```

```python
df['type name']=df['Iris type'].apply(convert)
```

```python
df.head()
```

**A-> Plot bar chart to show the frequency of each class label in the data.**

```python
df.groupby('type name').count()
```

```python
plt.figure(figsize=(3,4))
sns.countplot(x='type name',data=df)
```

__B -> Draw a scatter plot for Petal width vs sepal width.__

```python
df.columns
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='petal width (cm)',y='sepal width (cm)',data=df,hue='type name')
plt.legend(bbox_to_anchor=(1.35,0.9))
```

<h3>C -> Plot density distribution for feature petal length.</h3>

```python
sns.histplot(x='petal length (cm)',data=df,hue='type name',multiple='stack',palette='pastel')
```

**D -> Use a pair plot to show pairwise bivariate distribution in the Iris Dataset**


```python
sns.pairplot(df.drop('Iris type',axis=1),corner=1,hue='type name',palette='Dark2')
```
