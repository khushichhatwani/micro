<h1>Write a program in Python using Pandas to perform the following:</h1>
<p>a. Calculate and display familywise gross monthly income.</br>
b. Calculate and display the member with the highest monthly income in a family.</br>
c. Calculate and display monthly income of all members with income greater than Rs. 60000.00.</br>
d. Calculate and display the average monthly income of the female members in the Shah family.</p>

```python
df.describe()
```

```python
df['Monthly Income (Rs.)']=df['Monthly Income (Rs.)'].apply(lambda x:x.replace(",",""))
```

```python
df
```

```python
df['Monthly Income (Rs.)']=df['Monthly Income (Rs.)'].apply(pd.to_numeric)
```

```python
df
```

```python
df.describe()
```

**a. Calculate and display familywise gross monthly income.**

```python
df.groupby(by="Name").sum(numeric_only=1)
```

**b. Calculate and display the member with the highest monthly income in a family.**

```python
df.groupby(by="Name").max()
```

**c. Calculate and display monthly income of all members with income greater than Rs. 60000.00.**

```python
df.groupby(by="Name").apply(lambda x:x[x['Monthly Income (Rs.)']>60000.00])
```

**d. Calculate and display the average monthly income of the female members in the Shah family.**

```python
df.groupby(by="Name").apply(lambda x:x[x['Gender']=='Female'])
```

```python
df.groupby(by="Name").apply(lambda x:x[x['Gender']=='Female'].mean(numeric_only=1))
```
