```python
from datetime import datetime
```

```python
date_format = '%Y-%m-%d %H:%M:%S'
df["Formatted Date"]=df["Formatted Date"].apply(lambda x:datetime.strptime(x[:-10], date_format))
```

# d. Split a dataset to group by two columns and then sort the aggregated results within the groups.

```python
res=df.groupby(['Precip Type','Summary']).mean(numeric_only=1)['Humidity']
res
```

```python
res.groupby(level=1).sum().sort_values()
```
